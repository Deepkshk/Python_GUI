### Employee Database Simulation using Python GUI and SQLite

```python
# import required libraries
from tkinter import *
import sqlite3
from tkinter import messagebox
import re
from random import randint,randrange
```

```python
# create  required database and tables
try:
    con=sqlite3.connect(database='employee_database.sqlite')
    cur=con.cursor()
    
    t1='create table all_employee(employee_id integer primary key autoincrement default 1000,first_name text not null,last_name text not null,designation text not null,department text not null,salary decimal (5,2) not null,work_experience decimal (2,2) not null default 0,mobile_no text not null check(length(mobile_no)==10),email_id text not null check(email_id like "%_@_%._%"),pwd text check(length(pwd==8)))'
    
    cur.execute(t1)
    
    con.commit()
    con.close()
    
    print('Table Created')
    
except Exception as e:
    print('Something went Wrong,',e)
```
