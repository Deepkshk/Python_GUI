```python
# import all required libraries
from tkinter import *
import pymysql
from tkinter import messagebox
import re
from random import randint,randrange
```

```python
# create mysql table for project
try:
    con=pymysql.connect(host='localhost',port=3306,user='root',password='mysql',database='myproject')
    cur=con.cursor()
    t='create table all_employee(employee_id int primary key auto_increment,first_name varchar(20) not null,last_name varchar(20) not null,designation varchar(20) not null,department varchar(20) not null,salary float(9,2) not null,work_experience float(3,1) not null default 0,mobile_no varchar(10) not null,email_id varchar(20) not null,pwd varchar(10) not null)'
    cur.execute(t)
    con.commit()
    con.close()
    print('Table Created')
except Exception as e:
    print('Something went Wrong, ',e)
```
